```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path
```

```python
# generate 100 random samples
x = np.random.uniform(low=0.0, high=2.0, size=100)
y = np.random.uniform(low=0.0, high=2.0, size=100)
points = np.vstack((x, y)).T
```

```python
# define the nodes of polygon
coord = [[1,1], [2,1], [2,2], [1,2], [0.5,1.5]]
coord.append(coord[0]) #repeat the first point to create a 'closed loop'
xs, ys = zip(*coord) #create lists of x and y values
```

```python
# find samples inside and outside the polygon
p = Path(coord)
index = p.contains_points(points)
xi, yi = zip(*points[index])
xo, yo = zip(*points[~index])
```

```python
# plot all points and fill the polygon
plt.scatter(xi, yi, s=10, c='r')
plt.scatter(xo, yo, s=10, c='k')
plt.fill(xs, ys, facecolor='cyan', edgecolor='k', alpha=0.2)
```
